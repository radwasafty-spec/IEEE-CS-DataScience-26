```python
# import numpy library
import numpy as np
```

```python
# Create a 3x3 identity matrix and multiply by 9 to make diagonal = 9
matrix = np.eye(3) * 9

# Print result
print(matrix)
```

```python
# Create 7x7 matrix of ones
matrix = np.ones((7,7))

print(matrix)
```

```python
# Create a 5x5 hole of zeros inside the matrix
matrix[1:6, 1:6] = 0

print(matrix)
```

```python
# Place number 5 in the center of the matrix
matrix[3,3] = 5

print(matrix)
```

```python
# Ask user for n
n = int(input("Enter n: "))
```

```python
# Create array using broadcasting
arr = np.arange(1, n+1)
arr = np.tile(arr, (n,1))

print(arr)
```

```python
# Generate random 5x5 matrix from 0 to 100
matrix = np.random.randint(0, 101, (5,5))

print(matrix)
```

```python
# Find maximum value in each row
print(matrix.max(axis=1))
```

```python
# Find sum of each column
print(matrix.sum(axis=0))
```

```python
# Calculate overall mean
print(matrix.mean())
```

```python
# Boolean masking: values > 50 become 1, others 0
new_matrix = (matrix > 50).astype(int)

print(new_matrix)
```

```python
# Create array from 1 to 30 and reshape to 6x5
arr = np.arange(1,31).reshape(6,5)

print(arr)
```

```python
# Extract 2x2 square [[11,12],[16,17]]
print(arr[2:4, 0:2])
```

```python
# Extract elements [2, 8, 14, 20]
print(arr[[0,1,2,3], [1,2,3,4]])
```

```python
# Get rows with even indices (0,2,4)
print(arr[::2])
```

```python
# Create scores array
scores = np.array([45, 88, 92, 30, 50, 76, 99, 12, 65])

print(scores)
```

```python
# Print passing grades (>= 50)
print(scores[scores >= 50])
```

```python
# Check if any student got 100
print(np.any(scores == 100))
```

```python
# Check if all students passed
print(np.all(scores >= 50))
```
