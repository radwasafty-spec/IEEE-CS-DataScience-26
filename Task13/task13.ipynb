### Step 11. Which was the most-ordered item?

```python
c = chipo.groupby('item_name').sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```

### Step 12. For the most-ordered item, how many items were ordered?

```python
c = chipo.groupby('item_name').sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)['quantity']
```

### Step 13. What was the most ordered item in the choice_description column?

```python
c = chipo.groupby('choice_description').sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```

### Step 17. How many products cost more than $10.00?

```python
# أولاً: بنشيل علامة الـ $ ونحول السعر لرقم (float)
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:-1]))

# ثانياً: بنفلتر ونحسب العدد
products_over_10 = chipo[chipo['item_price'] > 10]
print(len(products_over_10))
```
